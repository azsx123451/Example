```python
#KNeighborsClassifier
#iris資料集
import numpy as np
from sklearn import datasets
from sklearn.cross_validation import train_test_split
from sklearn.neighbors import KNeighborsClassifier

iris = datasets.load_iris()
iris_x = iris.data
iris_y = iris.target
#print(iris_x[:,:])
#print(iris_y)

x_train, x_test, y_train, y_test = train_test_split(
    iris_x, iris_y, test_size=0.5)
#print(y_train)
knn =KNeighborsClassifier()
knn.fit(x_train,y_train)
print(knn.predict(x_test))
print(y_test)
```

```python
from sklearn import datasets
from sklearn.linear_model import LinearRegression

loaded_data = datasets.load_boston()
data_X = loaded_data.data
data_y = loaded_data.target

model = LinearRegression()
model.fit(data_X, data_y)

print(model.coef_)  #輸出模型的斜率
print(model.intercept_)  #輸出模型的截距
print(model.get_params())  #取出定義的參數
print(model.score(data_X, data_y))  # R^2 coefficient of determination  模型信賴程度
```

```python
from sklearn import preprocessing
import numpy as np
from sklearn.model_selection import train_test_split  #將資料分割成train與test模組
from sklearn.datasets.samples_generator import make_classification  #生成適合做classification的模組
from sklearn.svm import SVC  #Support Vector Machine中的Support Vector Classifier
import matplotlib.pyplot as plt  #數據視覺化

#生成具有兩種特徵屬性的1000筆資料   
#n_fratures : 總特徵數目  n_informative: 有意義的特徵數目  n_redundant : 產生有意義特徵的隨機線性組合 
#n_clusters_per_class: 一個類群有幾個群組分布
x,y = make_classification(n_samples=1000, n_features=2 , n_redundant=0, n_informative=2,
                           random_state=22, n_clusters_per_class=1, scale=100)


plt.scatter(x[:,0],x[:,1],c=y)
plt.show()
x = preprocessing.scale(x)
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.1)
clf = SVC()
clf.fit(x_train,y_train)
print(clf.score(x_test,y_test))

```

```python
from sklearn.datasets import load_iris
from sklearn.cross_validation import train_test_split
from sklearn.neighbors import KNeighborsClassifier

iris = load_iris()
x = iris.data
y = iris.target

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 4) #設有random_state準確度較高 
knn = KNeighborsClassifier(n_neighbors=10)  #切成十塊檢驗
knn.fit(x_train, y_train)

print(knn.score(x_test, y_test))
```

```python
# this is how to use cross_val_score to choose model and configs
from sklearn.datasets import load_iris
from sklearn.cross_validation import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cross_validation import cross_val_score
import matplotlib.pyplot as plt

iris = load_iris()
X = iris.data
y = iris.target

k_range = range(1,30)
k_scores = []
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors = k)
    #loss = -cross_val_score(knn, X, y, cv=10, scoring='mean_squared_error') # for regression
    scores = cross_val_score(knn,x,y,cv = 10,scoring = 'accuracy')  # for classification
    k_scores.append(scores.mean())

plt.plot(k_range,k_scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-Validated Accuracy')
plt.show()
```

```python
from sklearn.learning_curve import  learning_curve  #可視化學習過程
from sklearn.datasets import load_digits
from sklearn.svm import SVC
import matplotlib.pyplot as plt
import numpy as np

digits = load_digits()
X = digits.data
y = digits.target
train_sizes, train_loss, test_loss= learning_curve(
        SVC(gamma=0.0001), X, y, cv=10, scoring='mean_squared_error',
        train_sizes=[0.1, 0.25, 0.5, 0.75, 1])
train_loss_mean = -np.mean(train_loss, axis=1)
test_loss_mean = -np.mean(test_loss, axis=1)

plt.plot(train_sizes, train_loss_mean, 'o-', color="r",
             label="Training")
plt.plot(train_sizes, test_loss_mean, 'o-', color="g",
             label="Cross-validation")

plt.xlabel("Training examples")
plt.ylabel("Loss")
plt.legend(loc="best")
plt.show()


```

```python
from sklearn.learning_curve import  validation_curve  #可直觀看出改變模型參數有沒有overfitting的狀況，選擇更好參數的方法
from sklearn.datasets import load_digits
from sklearn.svm import SVC
import matplotlib.pyplot as plt
import numpy as np

digits = load_digits()
X = digits.data
y = digits.target

param_range = np.logspace(-5,-1,5)  #建立參數測試集
train_loss, test_loss = validation_curve(
        SVC(), X, y, param_name='gamma', param_range=param_range, cv=10,
        scoring='mean_squared_error')  #使用validation_curve快速找出参数對模型的影響
train_loss_mean = -np.mean(train_loss, axis=1)
test_loss_mean = -np.mean(test_loss, axis=1)

plt.plot(param_range, train_loss_mean, 'o-', color="r",
             label="Training")
plt.plot(param_range, test_loss_mean, 'o-', color="g",
             label="Cross-validation")

plt.xlabel("gamma")
plt.ylabel("Loss")
plt.legend(loc="best")
plt.show()
```
