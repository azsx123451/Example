```python
import re
import json
import requests
from bs4 import BeautifulSoup
i = 1
w = []
while i <= 10:
    url = "http://jhspt.com/case/page/{j}".format(j = i)
    a = requests.get(url)
    soup = BeautifulSoup(a.text, 'html5lib')
    c = soup.select('div.content h3 a')
    
    w.extend(c)  #導入陣列
    print(type(c))
    i+=1

print(w)
```

```python
str(w).split()[2][6:-1]  #用陣列位置取出網址
```
